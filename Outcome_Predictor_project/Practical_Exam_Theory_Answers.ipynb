# Practical Exam - Theory Answers

## 1. What is Regression? How is it different from Classification?

**Regression:** Predicts continuous numerical values (e.g., house price).

**Classification:** Predicts categories/classes (e.g., spam/not spam).

**Examples (India):**
- Predicting house prices in Mumbai.
- Predicting crop yield in Maharashtra.

| Regression | Classification |
|---|---|
| Continuous output | Categorical output |
| Example: House Price | Example: Loan Approved/Rejected |

## 2. Simple vs Multiple vs Ridge/Lasso Regression

**Simple Linear Regression:** Uses one independent variable.

**Multiple Linear Regression:** Uses two or more independent variables.

**Ridge Regression (L2):** Adds L2 penalty to reduce overfitting.

**Lasso Regression (L1):** Adds L1 penalty and performs feature selection.

## 3. Overfitting vs Underfitting

**Underfitting:** Model is too simple and performs poorly on training and testing data.

**Overfitting:** Model memorizes training data and performs poorly on unseen data.

**Regularization:** L1/Lasso removes less important features; L2/Ridge shrinks coefficients.

## 4. RMSE, MAE and R²

**MAE:** Average absolute error.

**RMSE:** Square root of average squared error; penalizes large errors more.

**R² Score:** Measures how well the model explains variance (1 is best).

**When to prefer RMSE?**
When large prediction errors are especially costly (e.g., house price prediction).

## 5. Cross Validation (K-Fold)

K-Fold Cross Validation divides the dataset into K folds. Each fold is used once as the test set while the remaining folds are used for training.

**Advantages**
- Better use of data.
- More reliable evaluation.
- Reduces dependence on a single train-test split.
